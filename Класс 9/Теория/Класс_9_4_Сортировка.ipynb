# Сортировка

## Сортировка пузырьком

|Лучшее|В среднем|В худшем|
|--|--|--|
|O(n)|O(n^2)|O(n^2)|

**Сортировка пузырьком** — это метод сортировки массивов и списков путем последовательного сравнения соседних элементов и их обмена, если предшествующий оказывается больше последующего (при сортировке по возрастанию).



## Сортировка выбором
|Лучшее|В среднем|В худшем|
|--|--|--|
|O(n)|O(n^2)|O(n^2)|

Проходим по массиву в поисках максимального элемента. Найденный максимум меняем местами с последним неотсортированным элементом.

![alt text](https://habrastorage.org/webt/yt/cs/fz/ytcsfzyhzn9xy8opfyodmgz-a4u.gif)

### Двухсторонняя сортировка выбором
Проходим по массиву в поисках максимального **и минимального** элемента. Найденный максимум меняем местами с последним неотсортированным элементом, а минимум с первым неотсортированным.
![alt text](https://habrastorage.org/webt/jj/xn/kq/jjxnkqnbcbgwtqhxq9_p99kwmdi.gif)

## Сортирока слиянием
|Лучшее|В среднем|В худшем|
|--|--|--|
|O(n*log(n))|O(n*log(n))|O(n*log(n))|



## Быстрая сортировка
|Лучшее|В среднем|В худшем|
|--|--|--|
|O(n*log(n))|O(n*log(n))|O(n^2)|
