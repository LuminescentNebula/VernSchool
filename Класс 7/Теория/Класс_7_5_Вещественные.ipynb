# Вещественные числа

Один из типов данных с которым мы пока не работали - это вещественные числа. Вещественные числа - это числа с дробной частью. 

В Python этот тип называется `float`, от английского названия числа с плавающей запятой.

## Ввод
Напоминание, `input()` по умолчанию считывает строку, до этого нам нужно было рабоать с вводом чисел, поэтому мы преобразовывали строку в целочисленное значение с помощью функции `int()`. Для того, чтобы преобразовать строку в тип вещественного числа нужно использовать функцию `float()`.

```python
a=float(input())
print(a)
```

## Вычисления
Можно производить операции между целыми и вещественными числами:  
`int + int = int`  
`float + float = float`  
`int + float = float`  
Если в вычислении участвует вещественное число, то и тип результата будет тоже вещественный.

Также тип `float` получается в результате деления (не целочисленного, а обычного) даже если дробная часть равна нулю:

```python
print(4/2)
```

## Округления

По умолчанию Python выводит только 15 знаков после запятой или меньше, если все знаки дальше равны нулю:

```python
print(12.1234567890123456789)
print(12.000000000000002)
print(12.0000000000000002)
```

Но иногда нужно выводить больше или меньше знаков. Для этого можно использовать форматированный вывод:

```python
number = 12.1251
print("{:.2f}".format(number))
```

где `:.2f` - округление до двух знаков после запятой, меняя 2 на другое число вы будете менять количество знаков после запятой. В этом случае само число не изменяется, только вывод.

Если необходимо именно округлить значение числа, то для этого есть функция `round()`. Первым аргументов в нее передается число, а вторым количество знаков после запятой до которых нужно округлить. Если второй аругмент не указать, то будет округление до целого числа.

```python
number = 12.1251
print(number)
number = round(number,2)
print(number)
```

Все что больше .5 `round` округляет в большую сторону, все что меньше или равно - в меньшую:

```python
print(round(12.50))
print(round(12.51))
```
