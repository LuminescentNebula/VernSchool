# Полиморфизм

Полиморфизм - принцип ООП - это способность обьекта использовать методы производного класса, который не существует на момент создания базового. Этот принцип тесно связан с понятием абстракции.

С помощью полиморфизма можно разработывать гибкие системы, даже пока конкретные части не реализованы. Например, в большинстве систем пользовательского интерфейса (UI) все элементы наследуются от базового класса, который реализует общие для всех параметры, такие как размеры и положение.

Можно описать базовый класс:

```python
from abc import ABC, abstractmethod

class Component(ABC): #Создаем абстрактный класс
    __width: float
    __height: float
    __x:float
    __y:float
    __name: str

    def get_width(self) -> float:
        return self.__width

    def get_height(self) -> float:
        return self.__height

    def get_x(self) -> float:
        return self.__x
    
    def get_y(self) -> float:
        return self._y

    @abstractmethod
    def come_on_do_something(self) -> str:
        return "Привет, меня зовут" + self.__name
```

И описывать логику для обработки всех будущих компонентов:

```python
components: list[Component] =[]

for i in components:
    print(i.come_on_do_something())
```

Дальше можно создать компоненты с самыми разными методами и данными, но их всех можно будет обрабатывать с помощью кода сверху. В список `components` будут добавляться компоненты наследники, в которых абстрактный класс будет реализован по своему.
