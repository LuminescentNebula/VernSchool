# Agile


Классическая методология управлением проектов называется Waterfall(Водопад). По ней вы сначала планируете, анализируете, проектируете, разрабатываете, тестируете, выпускаете и поддерживаете потом пока ваш проект не закроется. Такая методолгия не позволяет вернуться назад, чтобы изменить или дополнительно уточнить что-то.

 

Agile - это «гибкий подход» к управлению проектами: работать короткими итерациями, по чуть‑чуть, получать обратную связь и менять план по ходу дела.

В феврале 2001 года группа специалистов собралась в американском курортном месте и сформулировала Agile Manifesto. В нем они обозначили четыре ценности:

- Люди и взаимодействия важнее процессов и инструментов,
- Работающий продукт важнее исчерпывающей документации,
- Сотрудничество с заказчиком важнее согласования условий контракта,
- Готовность к изменениям важнее следования плану.

Манифест — это не набор правил, а сборник приоритетов. Agile позволил переосмыслить, как строятся проекты: вместо того чтобы посвятить месяцы подготовке документации, команды начали выпускать рабочие версии продукта итерациями, получать обратную связь и корректировать курс по мере необходимости

❓ Вопрос 3

Agile чаще всего ассоциитируется стакими цикличными структурами, поскольку в отиличии от классической методологии здесь процесс разработки итеративный. Часто при разработки по Agile в конце каждого цикла уже есть рабочий продукт, который можно выпустить, собрать отзывы и в следующем цикле улучшить и доделать.

## Scrum

Такие циклы обычно реализуются с помощью фреймворка Scrum.




❓ Вопрос 4


Спринт — это короткий временной интервал, в течение которого scrum-команда выполняет заданный объем работы.   

Бэклог - список задач. Есть Бэклог продукта - вообще все задачи, и бэклог спринта.  

Скрам мастер - это специалист, который помогает команде эффективно работать по принципам фреймворка Scrum, выступая как и фасилитатор. Это не начальник, не командир, не главный специалист по проекту, который все знает, это человек, который работает с процессами внутри команды


❓ Вопрос 5

В каждом спринте есть встречи - одна большая в начале и совсем маленькие (буквально на пару минут) каждый день. На таких встречах происходит обсуждение и распределение задач. Важная часть этого процесса - оценка сложности.

## Scrum Poker
Скрам покер (Покер планирования) - командная игра для оценки сложности задач или объема работы.

1) У каждого участника есть карточки с оценкой
2) Участники выбирают в закрытую карточку и одновременно открывают их для команды
3) В идеале у всех должна быть одна и та же оценка. Если расхождение большое, то скрам мастер просит участников с расхождениями высказать свое мнение почему они так оценивают задачу.
4) Карточки повторно не используются.

Скрам мастер записывает результаты по каждому заданию и подводит по каждому итог - среднее значение сложности.
