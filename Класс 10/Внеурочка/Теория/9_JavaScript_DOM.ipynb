# Объектная Модель Документа
Объектная Модель Документа (DOM) является программным интерфейсом для HTML. 

Используя объектную модель JavaScript может создавать динамические HTML:
- JS может изменять HTML элементы на странице
- JS может изменять атрибуты HTML элементов на странице
- JS может изменять CSSS стили на странице
- JS может удалять существующие HTML элементы и атрибуты
- JS может добавлять новые HTML элементы и атрибуты
- JS может реагировать на HTML события на странице
- JS может создавать события HTML на странице 

## Элементы

### Получение элементов в JavaScript

`document` - глобальная переменная текущей странице, содержит в себе всю страницу.  
Эти же методы можно использовать в любом элементе для поиска внутри.

| Селектор| JS |
|-----------------------|----|
|#paragraph | `document.getElementById("paragraph") //Один элемент`|
|.class | `document.getElementsByClassName("class") //Список элементов`|
|div | `document.getElementsByTagName("div") //Список элементов`|
| любой селектор как в `CSS` | `document.querySelector() //Первый элемент`<br>`document.querySelectorAll() //Список элементов` |

## События



### Обработка событий
```js
element.onclick = function() { ... } //функция
element.onclick = functionName //Название функции
```

`onclick` соответсвует событию `click`. Остальные события имеют такие же названия, например `ondblclick` - `dbclick`, `onmousemove` - `mousemove`. Просто добавляется on.

ИЛИ
```js
element.addEventListener('click', function() { ... })
element.addEventListener('click', functionName)
```
`.addEventListner` позволяет добавлять несколько обработчиков для одного и того же события. Например вы можете создать несколько функций и добавлять их в элементы по одному или вместе в зависимости от необходимости.

ИЛИ прямо в HTML
```html
<button onclick="functionName()">
<button onclick="() =>{...}()">
```

Такой способ рекомендуется только на кнопках, потому что это смешивает логику и разметку и уменьшает гибкость. Лучше добавлять обработчики событий только внутри JavaScript, а HTML не изменять для этого.

### Параметры событий
Обработчики событий в JavaScript автоматически получают первым аргументом объект события `event`, который содержит всю информацию о событии: тип, текущий элемент (`currentTarget`), координаты, кнопка и т.д. в зависимости от события.

```js
button.addEventListener('click', (event) => {...});
```


### События мыши
- `click` - нажатие левой кнопкой мыши
- `dblclick` - двойное нажатие левой кнопкой мыши
- `contextmenu` - нажатие правой кнопкой мыши
- `mouseover` / `mouseout` – когда мышь наводится на / покидает элемент.
- `mousedown` / `mouseup` – когда нажали / отжали кнопку мыши на элементе.
- `mousemove` – при любом движении мыши.
#### Параметры событий мыши
|Поле|Описание|
|---|---|
|`target`| Цель, элемент который реагирует на событие|
|`altKey`| `true`, если кнопка alt была зажата|
|`ctrlKey`| `true`, если кнопка ctrl была зажата|
|`shiftKey`| `true`, если кнопка shift была зажата|
|`metaKey`| `true`, если кнопка windows/command была зажата |
|`button`| Номер кнопки, которая была нажата<br>0 - ЛКМ<br>1 - СКМ<br>2 - ПКМ<br>3 - Четвертная кнопка (назад)<br>4 - Пятая кнопка (вперед) |
|`buttons`| Номера кнопок, которые были нажаты |
|`clientX`<br>`clientY`| Позиция мыши в пределах viewport |
|`x`<br>`y`| Позиция мыши в пределах viewport. Тоже самое что и `clientX`/`clientY` |
|`movementX`<br>`movementY`| Позиция мыши относительно последнего вызова mousemove |
|`offsetX`<br>`offsetY`| Позиция мыши относительно элемента |
|`pageX`<br>`pageY`| Позиция мыши относительно всей страницы |
|`relatedTarget`| Вторичная цель, обычно элемент из которого пришел курсор (`mouseenter`,`mouseover`) или элемент в который ушел курсор (`mouseleave`,`mouseout`) |

Пример:
```js
button.addEventListener('click', (event) => {
    console.log("Координаты: "+event.clientX+" "+event.clientY)
});
```


### События клавиатуры
- `keydown`/`keyup` – когда пользователь нажимает / отпускает клавишу.
#### Параметры событий клавиатуры
|Поле|Описание|
|---|---|
|`target`| Цель, элемент который реагирует на событие|
|`key`| Номер кнопки, которая была вызвала событие |
|`altKey`| `true`, если кнопка alt была зажата|
|`ctrlKey`| `true`, если кнопка ctrl была зажата|
|`shiftKey`| `true`, если кнопка shift была зажата|
|`metaKey`| `true`, если кнопка windows/command была зажата |
|`repeat`| `true`, если кнопка уже давно зажата и сработал автоповтор |

### Другие события
Событий очень много, буквально на каждый чих и на каждое дествие на любых устройствах. Например: 

- `scroll` - когда страница или элемент был пролистан  
- `transitionend` - когда transition закончил свое исполнение  
- `focus`/`blur` - когда элемент попал в фокус/вышел из фокуса (например при выборе элементов на странице с помощью клавитуры - TAB)  
- `copy` - когда происходит копирование
- `wheel` - когда двигается колесико мыши
