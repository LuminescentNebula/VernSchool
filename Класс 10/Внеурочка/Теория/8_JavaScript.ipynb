### Типизация
И JS и Python имеют динамическую типизацию, то есть вы не пишите при объявлении переменной какой у нее тип, но в Python типизация строгая, а в JS слабая.  
Строгая типизация не допускает операции между переменными разных типов

````python
a=3
b="Строка"
print(a+b)
````

### Синтаксис
Есть 2 основных типа синтаксиса:
1) Python-подобный синтаксис - структурные блоки определяются выравниванием строк (число пробелов в начале)
2) С-подобный синтаксис - структурные блоки определяются парными элементами(фигурными скобками)

Python: 
```python
def sum(a,b):
  c = a + b
  return c
```

JavaScript: 
```js
function sum(a, b) {
    let c = a + b;
    return c;
}
```

Отступы в JS тоже используются, но не для функционала, а для красоты. Так код легче читать. Можно и без них, но будет некрасиво.

Фигурные скобки используются в тех же местах, в которых в Python использовались отступы: циклы, условия, функции, классы. Обратите внимание, что скобки ставятся в начале блока кода и в конце.



| Конструкция            | Python                                                                 | JavaScript                                                                 |
|-----------------------|------------------------------------------------------------------------|----------------------------------------------------------------------------|
| Условия               | `if условие:`<br>&emsp;`блок_кода`<br>`elif условие:`<br>&emsp;`блок_кода`<br>`else:`<br>&emsp;`блок_кода`  | `if (условие) {`<br>&emsp;`блок_кода`<br>`} else if (условие) {`<br>&emsp;`блок_кода`<br>`} else {`<br>&emsp;`блок_кода`<br>`}` |
| Циклы                 | `while условие:`<br>&emsp;`блок_кода`<br><br>`for item in iterable:`<br>&emsp;`блок_кода` | `while (условие) {`<br>&emsp;`блок_кода`<br>`}`<br><br>`for (значение до запуска; условие прерывания; изменение значения после совершения итерации) {`<br>&emsp;`блок_кода`<br>`}` |
| Функции               | `def имя(параметры):`<br>&emsp;`блок_кода`<br>&emsp;`return значение`  | `function имя(параметры) {`<br>&emsp;`блок_кода`<br>&emsp;`return значение`<br>`}`  |
|Лямбда-функции |`lambda args: выражение`|`const имя = (args) => выражение;`|
| Классы                | `class Имя:`<br>&emsp;`def __init__(self, args):`<br>&emsp;&emsp;`self.атрибут = значение`<br>&emsp;`def метод(self):`<br>&emsp;&emsp;`блок_кода` | `class Имя {`<br>&emsp;`constructor(args) {`<br>&emsp;&emsp;`this.атрибут = значение;`<br>&emsp;`}`<br>&emsp;`метод() {`<br>&emsp;&emsp;`блок_кода`<br>&emsp;`}`<br>`}` |
| Присваивание          | `x = значение`<br>`x, y = 1, 2`                                        | `const x = значение;`<br>`let x = значение;`<br>`var x = значение; //`<br>`x = значение;` |
| Исключения            | `try:`<br>&emsp;`блок_кода`<br>`except Exception as e:`<br>&emsp;`блок_кода`  | `try {`<br>&emsp;`блок_кода`<br>`} catch(e) {`<br>&emsp;`блок_кода`<br>`}`  |


`const` - неизменяемые константы.  
`let` - объявление переменной внутри блока `{}` (циклы, условия, функции, классы и т.д.) или глобально, если вы вне любых блоков.  
`var` - объявление внутри функциональной видимости (внутри функции, а не только блока `{}`).

**`var` считается устаревшим. Используйте `let` и `const`.**
