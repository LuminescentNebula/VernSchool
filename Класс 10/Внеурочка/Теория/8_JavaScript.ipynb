# JavaScript
Ни в коем случае не путаем язык Java и язык JavaScript. Это два абсолютно разных языка используемых для разных целей и имеющие похожие названия только из-за маркетингого хода второго.

JavaScript используется практически на каждом сайте для добавления интерактивности и обработки данных. В целях безопасного использования в JS есть ряд ограничений:
- Нельзя читать и записывать файлы на компьютер;
- Нельзя работать с системными функциями ОС;
- Нельзя работать с камерой и микрофоном без разрешения пользователя;
- Взаимодействовать с другими окнами и вкладками (если эти страницы не согласны на это);
- Получать данные с других сайтов и доменов (если эти страницы не согласны на это).

Благодаря этим ограничениям можно не бояться, что случайный сайт прочитает все ваши файлы на компьютере и переведет все ваши деньги из банка, который вы открыли в соседней вкладке.

### Синтаксис
Есть 2 основных типа синтаксиса:
1) Python-подобный синтаксис - структурные блоки определяются выравниванием строк (число пробелов в начале)
2) С-подобный синтаксис - структурные блоки определяются парными элементами(фигурными скобками)



```python
def sum(a,b):
    c = a + b
    return c
```

```js
function sum(a, b) {
  let c = a + b;
  return c;
}
```
