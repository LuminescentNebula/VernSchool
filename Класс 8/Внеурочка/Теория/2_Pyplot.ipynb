# Графики
Есть много различных видов графиков

Но все они сводятся к нескольким основным типам:  


В `matplotlib` для есть модуль `pyplot`, который используется для создания графиков. Обычно его импортирубт с коротким названием `plt`:

```python
import matplotlib.pyplot as plt
```

| Функция       | Тип графика |
| ------------- | ----------- |
| [`plt.pie(x, labels=[])`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html)     | Круговой    |
| [`plt.bar(x, y)`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar)     | Столбчатый  |
| [`plt.hist(x, y)`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html#matplotlib.pyplot.hist)    | Гистограмма |
| [`plt.plot(x, y)`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)    | Линейный    |
| [`plt.scatter(x, y)`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter) | Точечный    |
