# Jupyter Notebook
## Описание
Чтобы писать на Python можно использовать интерпретатор в интерактивном режиме, который выполняет строки кода как только вы их вводите



или можно создавать файлы с кодом и исполнять его целиком, но тоже построчно, посколько Python - интерпретируемый язык.

>    Интеретируемые языки - интерпретатор переводит команды в машиночитаемую форму на лету строка за строкой
>    
>    Компилируемые языки - исходный код целиком преобразуется в машинный код компилятором до запуска программы

Еще один способ, который используют работе с Python - Jupyter Notebook (JN). JN состоит из ячеек, каждая из которых может содержать запускаемый код или язык разметки (текст). В JN удобно показывать графики, изображения и визуализировать данные, некоторые библиотеки, например pandas, оптимизированы именно под работу в Jupyter Notebook.

При работе в блокноте (Notebook) вы можете созадавать много ячеек с кодом на Python и запускать их по отдельности. Для их выполнения нужно запустить ядро Python и пока оно работает все ваши переменные и функции будут сохраняться в пределах блокнота. То есть вы можете объявить переменную в одной ячейке, а потом вызывать её в другой. Причем расположение ячеек не имеет значение, только порядок в котором вы их запускали.

Этот материал тоже написан внутри Jyputer Notebook.

Пример: 

```python
# Вызов переменной a
# Для вывода не обязательно писать print, а можно просто вызвать переменную или данные, которые вы хотите вывести, в конце ячейки

# Слева указан порядок в котором запускали ячейки
# Хотя эта ячейка стоит первой, но я запустил её второй, потому что я сделал это вручную
a
```

```python
# Эта ячейка была запущена первой
a = 7 
```

```python
# А эта третьей
a*2
```

Может показаться сложным, но на самом деле нужно просто понимать, что при запуске вы не начинаете с нуля, а продолжаете со всеми теми переменными и значениями в них, которые были раньше, если вы их не перезаписываете. Это понимание нужно, чтобы вы потом не гадали почему у вас тот же результат, хотя вы запустили заново. 

Если необходимо начать полностью сначала, то можно перезапустить ядро, либо очистить переменные.

## Установка и открытие
Для скачивания Jupyter Notebook нужно открыть командную строку, на Windows - Win+R и ввести "cmd"  
  

На Linux Ctrl + Alt + T.

Python уже должен быть у вас скачен, дальше нужно с помощью менеджера пакетов скачать JN. В зависимости от версии нужно ввести одну из двух комманд в командную строку:  
`pip install notebook`  
`pip3 install notebook`

После скачивания можно будет ввести команду `jupyter notebook` и после небольшой загрузки в браузере откроется страница, где можно будет выбрать папку, в которой вы хотите хранить блокноты и создать новый блокнот. При закрытии командной строки редактор не закроется автоматически, но перестанет работать.

\#\# Заголовок второго уровня

Для перехода на новую строку без отступа между абзацами - поставить два пробела в конце предыдущего абзаца.
