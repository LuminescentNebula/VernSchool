## Задание

```python
#Категории
categories = ['Электроника', 'Одежда', 'Игрушки', 'Спорт', 'Книги']
#Количество заказов в каждой категории соответсвенно
values_by_categories = [46, 44, 43, 36, 31]
#Месяцы
months = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 
              'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
#Количество заказов по месяцам
values_by_months = [14, 21, 15, 20, 18, 16, 15, 10, 19, 14, 21, 17]
#Средняя сумма заказа по месяцам
avg_by_months = [7023.36, 8514.48, 7005.8, 8610.8, 9082.94, 8002.19, 
            7603.07, 6841.2, 8874.84, 7110.86, 9030.19, 6082.24]
```

### Задание 1
Покажите, какую долю от общего количества заказов составляет каждая категория товаров. Визуализируйте структуру заказов по категориям в процентах.

```python
plt.pie(values_by_categories, labels=categories, autopct='%1.1f%%')
plt.title('Задание 1: Структура продаж по категориям')

plt.show()
```


### Задание 2
Сравните количество заказов по месяцам. В каких месяцах было больше всего заказов, а в каких меньше всего?

```python
plt.bar(range(len(months)), values_by_months, color='steelblue')
plt.xticks(range(len(months)), months, rotation=45, ha='right')
plt.ylabel('Количество заказов')
plt.title('Задание 2: Количество заказов по месяцам')

plt.show()
```


### Задание 3
Покажите, как изменяется средняя сумма заказа по месяцам в течение года. Есть ли тенденция роста или падения?


```python
plt.plot(range(len(months)), avg_by_months, marker='o', linewidth=2, markersize=6, color='green')
plt.xticks(range(len(months)), months, rotation=45, ha='right')
plt.ylabel('Средняя сумма (руб)')
plt.title('Задание 3: Динамика средней суммы заказа')

plt.show()
```
